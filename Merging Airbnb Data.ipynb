```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)    
os.chdir("/Users/Philip/Downloads")  
df_reviews=pd.read_csv("cleaned_reviews_airbnb.csv")
df_listings=pd.read_csv("cleaned_listings_airbnb.csv")
df_calendar=pd.read_csv("cleaned_calendar_airbnb.csv")
```

```python
ax_price = sns.catplot(x="Year",y='price',hue='Month', kind='box',showfliers=False,data=df_listings)
#ax.legend(loc='upper right')
ax_price.fig.suptitle('Listing Prices Box Plot')

```

```python
ax_listnum = sns.catplot(x="Year",hue='Month', kind='count',data=df_listings)
#ax.legend(loc='upper right')
ax_listnum.fig.suptitle('Listing Counts')
plt.savefig('listingcounts.jpg')
```
